```python
def standardize_party_name(party):
    synonyms = [
        {'synonym' : 'Christen Democratisch Appel (CDA)', 'standardized' : 'CDA'},
        {'synonym' : 'Partij van de Arbeid (P.v.d.A.)', 'standardized' : 'PvdA'},
        {'synonym' : 'Partij van de Arbeid (P.v.d.A)', 'standardized' : 'PvdA'},
        {'synonym' : 'Democraten 66 (D66)', 'standardized' : 'D66'},
        {'synonym' : 'SP (Socialistische Partij)', 'standardized' : 'SP'},
        {'synonym' : 'Staatkundig Gereformeerde Partij (SGP)', 'standardized' : 'SGP'},
        {'synonym' : 'GROENLINKS', 'standardized' : 'GroenLinks'},
        {'synonym' : 'Groenlinks', 'standardized' : 'GroenLinks'},
        {'synonym' : 'PVV (Partij voor de Vrijheid)', 'standardized' : 'PVV'},
        {'synonym' : 'GemeenteBelangen', 'standardized' : 'Gemeentebelangen'},
        {'synonym' : 'Gemeente Belangen', 'standardized' : 'Gemeentebelangen'},
        {'synonym' : 'Partij Gemeente Belangen', 'standardized' : 'Gemeentebelangen'},
        {'synonym' : 'Combinatie Gemeentebelangen', 'standardized' : 'Gemeentebelangen'},
        {'synonym' : 'ChristenUnie/SGP', 'standardized' : 'SGP/ChristenUnie'},
        {'synonym' : 'ChristenUnie-SGP', 'standardized' : 'SGP/ChristenUnie'},
        {'synonym' : 'SGP-ChristenUnie', 'standardized' : 'SGP/ChristenUnie'},
        {'synonym' : 'Christenunie', 'standardized' : 'ChristenUnie'},   
        {'synonym' : 'PVDA/GroenLinks', 'standardized' : 'PvdA/GroenLinks'},
        {'synonym' : 'PvdA/GROENLINKS', 'standardized' : 'PvdA/GroenLinks'},
        {'synonym' : 'PvdA-GROENLINKS', 'standardized' : 'PvdA/GroenLinks'},
        {'synonym' : 'PvdA-Groenlinks', 'standardized' : 'PvdA/GroenLinks'},
        {'synonym' : 'PvdA-GroenLinks', 'standardized' : 'PvdA/GroenLinks'},
        {'synonym' : 'Lokale Politieke Federatie Westland', 'standardized' : 'LPF Westland'},
        {'synonym' : 'GemeenteBelang Westland (GBW)', 'standardized' : 'GemeenteBelang Westland'},
        {'synonym' : '', 'standardized' : ''},
        {'synonym' : '', 'standardized' : ''},
        {'synonym' : '', 'standardized' : ''}        
    ]
    
    for obj in synonyms:
        if party == obj['synonym']:
            party = obj['standardized']
            
    return party

def standardize_gender(gender):
    if gender == 'm':
        return 'male'
    elif gender == 'v':
        return 'female'
    else:
        return 'unknown'
    
def guess_gender(title):
    titles = title.lower().split(' ')
    male_titles = ['dhr.', 'jhr.']
    female_titles = ['mw.']

    if len(list(set(titles) & set(male_titles))) > len(list(set(titles) & set(female_titles))):
        return 'm'
    elif len(list(set(titles) & set(female_titles))) > len(list(set(titles) & set(male_titles))):
        return 'v'
    else:
        return 'unknown'
```

```python
def count_first_names(politicians):
    counter = 0
    for obj in politicians:
        if len(obj['first_name']) > 1:
            counter += 1
    print(counter)
```

```python
import csv

politicians_almanak = []
politicians_candidates = []

with open('../data_resources/politicians/politicians_almanak.csv') as csv_file:
        politicians = csv.DictReader(csv_file, delimiter=',')
        
        for person in politicians:
            politicians_almanak.append(person)
        
with open('../data_resources/politicians/kandidaten-gr2018.csv') as csv_file:
        politicians = csv.DictReader(csv_file, delimiter=',')
        
        for person in politicians:
            politicians_candidates.append(person)

```

```python
import collections

col_almanak = []
for obj in politicians_almanak:
    col_almanak.append('{}{} {} {}'.format(obj['initials'], obj['last_name'], obj['party'], obj['municipality']))
    
collections.Counter(col_almanak).most_common()
```

```python
col_candidates = []
for obj in politicians_candidates:
    col_candidates.append('{}{} {} {}'.format(obj['Voorletters'], obj['Achternaam'], obj['Partij'], obj['Gemeente']))
    
collections.Counter(col_candidates).most_common()
```

```python
count_first_names(politicians_almanak)
            
for politician in politicians_almanak:
    for candidate in politicians_candidates:
        count = 0
        
        if candidate['Voorletters'].lower() == politician['initials'].lower() \
        and candidate['Achternaam'].lower() == politician['last_name'].lower() \
        and candidate['Gemeente'].lower() == politician['municipality'].lower() \
        and standardize_party_name(candidate['Partij']).lower() == standardize_party_name(politician['party']).lower():
            count += 1
            politician['first_name'] = candidate['Voornaam']
            candidate['party'] = standardize_party_name(candidate['Partij'])
            
            if guess_gender(politician['title']) in ['m', 'v']:
                politician['gender'] = standardize_gender(guess_gender(politician['title']))
            else:
                politician['gender'] = standardize_gender(candidate['Geslacht'])

            if count > 1:
                print('politician: {}'.format(politician))
                print('candidate: {}'.format(candidate))
            
count_first_names(politicians_almanak)
print(len(politicians_almanak))
```

```python
import csv

people_old = []
parties_old = []

people_new = []
parties_new = []

with open('../data_resources/politicians/politicians.csv') as csv_file:
        politicians_old = csv.DictReader(csv_file, delimiter=',')
        
        for person in politicians_old:
            people_old.append('{} {}'.format(person['initials'], person['last_name']))
            parties_old.append(person['party'])
            

with open('../data_resources/politicians/kandidaten-gr2018.csv') as csv_file:
        politicians_new = csv.DictReader(csv_file, delimiter=',')


        for person in politicians_new:
            people_new.append('{} {}'.format(person['Voorletters'], person['Achternaam']))
            parties_new.append(person['Partij'])
            
print(len(people_old))
print(len(people_new))
```

```python
positives = []
negatives = []

for obj in people_old:
    if obj in people_new:
        positives.append(obj)
    else:
        negatives.append(obj)
```

```python
print(len(positives))
print(len(negatives))
print(len(positives) / (len(positives) + len(negatives)))
```

# Make new list of politicians

```python
people_complete = []

# new_format = [
#     'municipality',
#     'party',
#     'initials',
#     'first_name',
#     'last_name',
#     'gender',
#     'role'
# ]

with open('../data_resources/politicians/kandidaten-gr2018.csv') as csv_file:
        politicians = csv.DictReader(csv_file, delimiter=',')
        for person in politicians:
            
            object_already_in_list = False
            for obj in people_complete:
                if obj['initials'] == person['Voorletters'] and obj['last_name'] == person['Achternaam']:
                    object_already_in_list = True
               
            if not object_already_in_list:
                people_complete.append({
                    'municipality' : person['Gemeente'],
                    'party' : standardize_party_name(person['Partij']),
                    'initials' : person['Voorletters'],
                    'first_name' : person['Voornaam'],
                    'last_name' : person['Achternaam'],
                    'gender': standardize_gender(person['Geslacht']),
                    'role' : ''
                })
                
print('First one done')

with open('../data_resources/politicians/politicians.csv') as csv_file:
        politicians = csv.DictReader(csv_file, delimiter=',')
        
        for person in politicians:
            
            object_already_in_list = False
            for obj in people_complete:
                if obj['initials'] == person['initials'] and obj['last_name'] == person['last_name'] and standardize_party_name(obj['party']) == standardize_party_name(person['party']):
                    object_already_in_list = True
            
            if not object_already_in_list:
                gender = guess_gender(person['title'])
                
                people_complete.append({
                    'municipality' : person['municipality'],
                    'party' : standardize_party_name(person['party']),
                    'initials' : person['initials'],
                    'first_name' : person['given_name'],
                    'last_name' : person['last_name'],
                    'gender': standardize_gender(gender),
                    'role' : person['role']
                })
```

```python
# for obj in people_complete:
#     print(obj)

print(len(people_complete))
```

```python
import hashlib

def string2numeric_hash(text):
    text = text.encode('utf-8')
    return int(hashlib.md5(text).hexdigest()[:10], 16)


for obj in people_complete:
    obj['system_id'] = string2numeric_hash(obj['initials'] + ' ' + obj['last_name'])
    
    obj['initials'] = obj['initials'].replace(",", ".")
    obj['initials'] = obj['initials'].replace("'", "")
    obj['initials'] = obj['initials'].strip()
    
    
people_complete[0]

with open('../data_resources/politicians/politicians_complete.csv', 'w') as csvfile:
    fieldnames = ['system_id', 'municipality', 'party', 'initials', 'first_name', 'last_name', 'gender', 'role']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for person in people_complete:
        writer.writerow(person)
```
