```python
import json
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

def find_most_similar_slug(full_topic_name, slugs):
    final_slug = ''
    slug_score = 0.2
    
    for slug in slugs:
        score = similar(full_topic_name, slug)
        
        if score > slug_score:
            slug_score = score
            final_slug = slug
    
    return final_slug
```

```python
file = '../data_resources/topics/vng_topics.json'
topics = json.load(open(file))

file = '../data_resources/topics/vng_training_cleaned_no_news.json'
training_data = json.load(open(file))

main_slugs = []
sub_slugs = []
for obj in training_data:
    
    if not obj['sub_topic'] in sub_slugs:
        sub_slugs.append(obj['sub_topic'])
        
    if not obj['main_topic'] in main_slugs:
        main_slugs.append(obj['main_topic'])


topics = topics['topics']
for topic in topics:
    full = topic['name']
    
    slug = find_most_similar_slug(full, main_slugs)
    topic['slug'] = slug
    
    print('full: {} \nslug: {}\n'.format(full, slug))
    
    for sub_topic in topic['sub']:
        full = sub_topic['name']
        slug = find_most_similar_slug(full, sub_slugs)
        sub_topic['slug'] = slug
        
        print('full: {} \nslug: {}\n'.format(full, slug))
        
result = {'topics' : topics}

with open('../data_resources/topics/vng_topics_slug.json', 'w') as outfile:
    json.dump(result, outfile)
```
